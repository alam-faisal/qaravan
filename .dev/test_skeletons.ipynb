```python
Lx = 4
assert index_from_coords(2, 1, Lx) == 6
assert index_from_coords(1, 2, Lx) == 9
assert coords_from_index(6, Lx) == (2, 1)
assert coords_from_index(9, Lx) == (1, 2)
assert not horiz_filter((1,4))
assert horiz_filter((1,2))
assert even_filter((0,1), vert=False, Lx=Lx)
assert not even_filter((1,2), vert=False, Lx=Lx)
assert even_filter((0,4), vert=True, Lx=Lx)
assert even_filter((1,5), vert=True, Lx=Lx)
assert not even_filter((4,8), vert=True, Lx=Lx)
assert not even_filter((5,9), vert=True, Lx=Lx)

Ly = 3 
assert len(nn_edges(Lx,Ly)) == 2*Lx*Ly - Lx - Ly
assert brickwall_skeleton(Lx=4, Ly=2, depth=1) == [
    (0, 1),
    (2, 3),
    (4, 5),
    (6, 7),
    (1, 2),
    (5, 6),
    (0, 4),
    (1, 5),
    (2, 6),
    (3, 7)]

num_gates = 4
assert len(all_skeletons(num_gates, Lx=Lx, Ly=Ly)) == (2*Lx*Ly - Lx - Ly)**num_gates
```
