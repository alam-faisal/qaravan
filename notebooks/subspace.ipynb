```python
from scipy.special import comb

def su2_subspace_dim(N,j): 
    return comb(N,int(N/2-j)) - comb(N, int(N/2-j)-1) #/(2*j+1)#/comb(N,int(N/2))

def u1_subspace_dim(N,k): 
    return comb(N,k)
```

```python
su2_subspace_dim(4, 2)   
```

```python
su2 = [su2_subspace_dim(n,2) for n in range(6,20,2)]
u1 = [u1_subspace_dim(n,2) for n in range(6,20,2)]

import matplotlib.pyplot as plt
import numpy as np 
plt.plot(np.arange(6,20,2), su2, label='su2, total spin = 2')
plt.plot(np.arange(6,20,2), u1, label='u1, filling = 2')
#plt.yscale('log')
plt.ylabel("subspace dimension")
plt.xlabel("system size")
plt.legend()
```
