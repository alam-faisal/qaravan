# Test Kagome lattice

```python
row_layout = [{'num_cells': 3, 'shift': 0.0}, 
             {'num_cells': 2, 'shift': 0.5}, 
             {'num_cells': 1, 'shift': 1.0}]
lat = KagomeLattice(row_layout, periodic_x=False, periodic_y=False)
lat.spruce([0,7,17])
lat.plot()
```

```python
row_layout = [{'num_cells': 1, 'shift': 0.0}, 
             {'num_cells': 2, 'shift': -0.5}, 
             {'num_cells': 3, 'shift': -1}]
sp = [0,1,9,11,16,17]
lat = KagomeLattice(row_layout, periodic_x=False, periodic_y=False)
lat.spruce(sp)
lat.plot()
```
