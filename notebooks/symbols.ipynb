```python
import sympy as sp

sigma_i = sp.eye(2)
sigma_x = sp.Matrix([[0, 1], [1, 0]])
sigma_y = sp.Matrix([[0, -sp.I], [sp.I, 0]])
sigma_z = sp.Matrix([[1, 0], [0, -1]])
```

```python
a = sp.symbols('a0:6', real=True)
zi = sp.kronecker_product(sigma_z, sigma_i)
iz = sp.kronecker_product(sigma_i, sigma_z)
xx = sp.kronecker_product(sigma_x, sigma_x)
yy = sp.kronecker_product(sigma_y, sigma_y)
xy = sp.kronecker_product(sigma_x, sigma_y)
yx = sp.kronecker_product(sigma_y, sigma_x)

H = a[0] * zi + a[1] * iz + a[2] * xx + a[3] * yy + a[4] * xy + a[5] * yx
U = sp.exp(-sp.I * H)
sp.pprint(U)
```
