```python
from qaravan.core import ParamGate, arr_constructor
import numpy as np
```

```python
class CRX(ParamGate):
    def __init__(self, indices, theta):
        super().__init__("CRX", indices, theta)

    def construct_matrix(self): 
        return arr_constructor([
            [1, 0, 0, 0],
            [0, np.cos(self.angles[0]/2), 0, -1j*np.sin(self.angles[0]/2)],
            [0, 0, 1, 0],
            [0, -1j*np.sin(self.angles[0]/2), 0, np.cos(self.angles[0]/2)]
        ])
```
